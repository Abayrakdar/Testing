Join & Merge

Inner Left Join is all rows from left table with matching rows from the right table.
While Outer Join is all rows from both tables.

```python
Employees = pd.DataFrame({'ID': [1, 2, 3, 4, 5],
    'Name': ['Sara', 'Katie', 'David', 'Hazal', 'Georgie'],
    'Address': [42, 38, 36, 41, 35],
    'Salary': [33000, 45000, 65000, 95000, 35000]
})
Employees
```

```python
Department = pd.DataFrame({
    'ID' :[1,3,5,7],
     'Course':['Sales','HR','Marketing','Finance']
})
Department
```

```python
# Inner Join that is all matching rows
inner_join_df=pd.merge(Employees, Department, on = "ID", how="inner" )
inner_join_df
```

```python
# Outer joins returns all rows from all 1st and 2nd dataframes.
outer_join_df=pd.merge(Employees, Department, on ="ID", how="outer")
outer_join_df
```

```python
# Left join> all rows from first table and matching frmo second.
left_join_df=pd.merge(Employees, Department, on = "ID", how="left")
left_join_df
```

```python
# Right join => gives all rows from the 2nd table and matching rows from 1st table.
right_join_df=pd.merge(Employees, Department, on="ID",how="right")
right_join_df
```

```python
# Concat function in Python is used to combine tables or dataframes together.
concat_df=pd.concat([Employees,Department],axis=1) # axis is set to equal 1 means comcat along the column
# while if axis is to 0 means concat along the rows.
concat_df
```
